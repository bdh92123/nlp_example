```python
from konlpy.tag import Kkma, Okt, Komoran, Mecab
from konlpy.utils import pprint
```

```python
def posToString(sentence, tagger):
    sen = ""
    for tup in tagger.pos(sentence):
        if tup[1] in tagger.tagset:
            pos = tagger.tagset[tup[1]]
        else:
            pos = tup[1]            
        sen += tup[0] + "/" + pos.replace(" ", "") + "+"        
    return sen[:-1]
```

```python
kkma = Kkma()
twit = Okt()
kmr = Komoran()
mecab = Mecab()
```

```python
sentences = ['환절기에는 일교차가 심하니 특히 감기를 조심하자.','나는 어제 밥을 먹었다.','파일 제출 양식을 잘 지켜주세요', '짙은 파란색은 깊은 느낌을 준다.', '종원이는 박사를 하기로 마음을 먹었다.',
            '사실 명언이기 위해서는 그 사람이 우리에게 한 말이 머릿속에 오래 기억되고 다시 생각 나야 한다.',
            '문제를 해결하는 능력도 중요하지만 새로운 문제를 발견하고 왜 문제인지 이해하는 것이 우선인 것을 명심하자.',
            '좋아하는 것은 그 대상이 나를 기쁘게 해주길 바라는 것이고 사랑하는 것은 나로 인해 그 대상이 기쁘기를 바라는 것이다.']
```

```python
for sen in sentences:
    pprint(posToString(sen, kkma))
    pprint(posToString(sen, twit))
    pprint(posToString(sen, kmr))
    pprint(posToString(sen, mecab))
    print()

```
